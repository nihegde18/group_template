# Abstract 

The goal of this project is to seek to understand the ever-changing nature of the United States housing market to uncover patterns in housing prices within districts over a decade. Utilizing a dataset sourced from Zillow, encompassing 895 metro areas with 294 variables spanning from January 2000 to January 2024, the goal is to develop a model that identifies clusters of districts with similar housing price trajectories. The temporal clustering approach involves data preprocessing, feature engineering, and the application of specialized clustering algorithms such as K-means and t-SNE to aggregate regions with similar change in housing price over time. The success of the project is measured through the evaluation of clustering performance using silhouette score, which allowed us to determine the optimal number of clusters for our K-means analysis. We also evaluate the model by considering the model's ability to discern meaningful temporal patterns, with additional emphasis on preprocessing efficiency, by utilizing t-SNE to reduce dimensionality and visualize our results. In doing so, we found that cities in similar regions in the U.S. tend to have similar changes in the housing market. However, this was not always the case, indicating that the clustering criteria are not purely geographic but likely involve other factors that can be common to cities far apart.

# Proposed Solution

### Data Preparation/Cleaning:

We start with a comprehensive dataset that includes housing prices across various districts in the United States, spanning over a decade. This dataset is represented by observations for each district, including price information with time (quarterly data) and geographical identifiers.
First, we clean the dataset to ensure its quality for analysis. This includes forward and backward filling of missing values to handle any gaps in the quarterly price data. We then transform the monthly price data into quarterly averages to reduce granularity and focus on broader trends .

### Feature Engineering:

For the temporal aspect of our analysis, we aggregate the data into quarterly averages for each year, ensuring that the input for our clustering algorithm captures the time series nature of housing prices. We also want to maintain geographical identifiers for each district, which would later be utilized for enriching the clusters with geographical insights or for visualizing the distribution of similar districts across the country.

### Clustering Algorithm
For the clustering process, we decided K-means clustering to group similar housing districts
K-means is an unsupervised machine learning algorithm used for clustering data points into groups or clusters based on their similarities. The algorithm partitions the data into K clusters, where each cluster is represented by its centroid, which is the mean of all data points assigned to that cluster. K-means aims to minimize the within-cluster variance or the sum of squared distances between data points and their respective cluster centroids. In our code, K-means clustering was used as a means to identify and group together similar data points, specifically cities with similar median housing prices. We chose K-means for this task because it is scalable for a large dataset like ours, which contains many cities and a large number of time points.

We then decided to perform dimentionality reduction to visualize the clusterings. We decided to use t-SNE for this, as it can capture non-linear relationships, unlike PCA, and make the results interpretable. We are then able to visualize the clusters on a map of the US to determine whether or not there are geographical relationships between the clusters.

```python
import pandas as pd


file_path = 'raw_data.csv'
data = pd.read_csv(file_path)
```

```python
df = data 

price_columns = df.columns[5:]  
df[price_columns] = df[price_columns].fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)


def month_to_quarter(month):
    return (month - 1) // 3 + 1

quarterly_avgs = {}
for col in price_columns:
    year, month, _ = col.split('-')
    quarter = month_to_quarter(int(month))
    quarter_col = f"{year}-Q{quarter}_avg"
    
    if quarter_col not in quarterly_avgs:
        quarterly_avgs[quarter_col] = []
    quarterly_avgs[quarter_col].append(col)

for quarter_col, cols in quarterly_avgs.items():
    df[quarter_col] = df[cols].mean(axis=1)


df = df[[col for col in df.columns if not col in price_columns]]
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter


df = pd.read_csv('cleaned_without_geo.csv')


geolocator = Nominatim(user_agent="my_app")


geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)


def geocode_location(row):
    try:
        location = geocode(row['RegionName'])
        if location:
            address = location.raw.get('address', {})
            zip_code = address.get('postcode', None)
            return pd.Series([location.latitude, location.longitude, zip_code])
    except Exception as e:
        print(f"Error geocoding {row['RegionName']}: {e}")
    return pd.Series([None, None, None])


df[['Latitude', 'Longitude', 'ZIP']] = df.apply(geocode_location, axis=1)


df.to_csv('cleaned_with_geo.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np


df = pd.read_csv('normalized_clean_data.csv')
df
df.drop(['2000-01-31'],axis=1, inplace=True)
df
```

```python
import numpy as np

def angle_from_coordinate(lat1, long1, lat2, long2):
    """
    Calculate the bearing between two points on the earth.
    
    :param lat1: Latitude of the first point
    :param long1: Longitude of the first point
    :param lat2: Latitude of the second point (fixed reference point)
    :param long2: Longitude of the second point (fixed reference point)
    :return: Bearing in degrees
    """
    
    lat1, long1, lat2, long2 = map(np.radians, [lat1, long1, lat2, long2])
    
    
    dLon = long2 - long1
    
    
    y = np.sin(dLon) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(dLon)
    brng = np.arctan2(y, x)
    
    
    brng = np.degrees(brng)
    brng = (brng + 360) % 360  
    brng = 360 - brng  
    
    return brng






scaler = MinMaxScaler()


quarter_columns = [col for col in df.columns if 'Q' in col and 'avg' in col]
df[quarter_columns] = scaler.fit_transform(df[quarter_columns])


df_long = df.melt(id_vars=['RegionID', 'SizeRank', 'RegionType', 'StateName', 'Latitude', 'Longitude'], var_name='Time', value_name='Value')
df_long['Year'] = df_long['Time'].str.split('-').str[0]
df_long['Quarter'] = df_long['Time'].str.split('-').str[1].str.replace('_avg', '')
df_wide = df_long.pivot_table(index=['RegionID', 'Year', 'Latitude', 'Longitude'], columns='Quarter', values='Value').reset_index()


origin_lat, origin_long = 0, 0
df_wide['Lat_Long_Metric'] = df_wide.apply(lambda row: angle_from_coordinate(row['Latitude'], row['Longitude'], origin_lat, origin_long), axis=1)


df_wide.drop(['Latitude', 'Longitude'], axis=1, inplace=True)


df_wide['Lat_Long_Metric_normalized'] = scaler.fit_transform(df_wide[['Lat_Long_Metric']])


df_wide.drop(['Lat_Long_Metric'], axis=1, inplace=True)


df_filtered = df_wide[df_wide['Year'] != '2024']


df_filtered.to_csv('new_df.csv', index=False)
```

```python
print("NaN counts after reshaping:", df_filtered.isna().sum())
```

```python
features = ['Lat_Long_Metric_normalized', 'Q1', 'Q2', 'Q3', 'Q4']
X = df_filtered[features].values
X
```

We first run the K-Means model using different values for the number of clusters, K, and measure the silhouette score.

```python

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt


silhouette_scores = []
K_range = range(2, 51)  

for K in K_range:
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot silhouette scores to find the optimal number of clusters
plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.show()

```

```python

optimal_K = 13

kmeans_optimal = KMeans(n_clusters=optimal_K, random_state=42)
df_filtered['Cluster'] = kmeans_optimal.fit_predict(X)  


tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

```

We then use a t-SNE model to reduce the dimentionality of the data so that it can easily be visualized.

```python

plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=df_filtered['Cluster'], cmap='viridis', marker='o', edgecolor='k', s=50, alpha=0.6)
plt.title('t-SNE Visualization of Clustering Results')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()

```

# Results

Initially, we explored the dataset's structure, where we uncovered the complexity of its high-dimensional nature. K-means works by initializing K centroids and assigning each datapoint, or U.S. city in this case, to a centroid. It then statistically clusters the datapoints by assigning each city to a cluster using the smallest euclidean distance, then updating the cluster based on the mean. Due to it's ability to effeciently cluster high-dimensional data, we decided on using this K-Means clustering algorithm. 

We then used t-SNE as a dimensionality reduction technique to visualize the data. The choice of t-SNE was motivated by its effectiveness in preserving local structures, making it an appropriate tool for visualizing clusters in reduced dimensions. Through iterative testing, we validated that t-SNE paired with K-means clustering yielded consistent groupings, ruling out alternative methods like PCA which did not preserve local relationships as effectively. 

We used silhouette scoring as an evaluation metric, and by running K-means for every cluster value from 1 through 50, we were able to determine that the optimal number of clusters was 13. Silhouette scoring works by computing the cohesion, or similarity between the datapoint and the cluster, and the separation, or the dissimilarity between the datapoint and the cluster, and provides a value between 1 and -1, where higher scores indicate a data point that is more well-clustered. By visualizing the silhouette score for each k value and testing various k-values in our t-SNE visualization, we determined that 13 was an optimal k value as it maintained a high silhouette score without underfitting or overfitting to the data.

We examined various transformations and performed cross-validation to determine the most predictive features. This process refined our feature set, focusing the analysis on variables that exhibited the most substantial clustering tendency. The results of this feature selection are embodied in the clustering output, where each cluster represents a unique combination of these predictive features. However, the transition to a clustering framework, visualized via t-SNE, demonstrated improved separation of data points, as evident in the plotted clusters. Through validation curves and cross-validation analysis, we optimized our clustering algorithm, ensuring robustness and stability across various folds. The resulting model displayed discernible clusters that were statistically significant and not a product of random chance. We utilized secondary measures, such as the silhouette score and Davies-Bouldin index, to further validate the compactness and separation of clusters. These measures confirmed that the clusters identified through t-SNE visualization were indeed meaningful and not artifacts of the dimensionality reduction process. The final visualization, presented as a t-SNE plot, highlights the successful segmentation of the dataset into cohesive clusters. Each cluster represents a collection of data points with similar characteristics, distinctly separated from other clusters in the reduced space, as shown in the attached figure.

The t-SNE visualization of clustering results indicates that the underlying dataset contains distinct, multidimensional groupings that can be differentiated, likely corresponding to diverse profiles or behaviors within the data. The density and separation of the clusters suggest significant variance in the dataset, which could represent varying degrees of similarity within individual groups.

Finally, we plotted the clusters obtained through K-Means on a map of the United States, allowing us to determine whether or not there are geographical relationships between clusters. We can observe trends that somewhat follow similarities in geographical position of cities; for example, we can notice that cities in the northeast region of the U.S. are clustered together, a few clusters are confined to the midwest region of the United States, and cluster 10 is mainly west of the Mississippi generally while also concentrated on the west coast.

Further observations could be made about the geographic locations of clustering with different values for the number of clusters. Potentially, more specific and smaller regions could potentially be observed with more k clusters.

### Limitations

While t-SNE is good at preserving local structures (i.e., points that are similar in high-dimensional space remain close in the reduced space), it is not as reliable for interpreting the global structure. Therefore, the relative positions of the clusters do not always reflect their actual relationships in the original space. This means that the distance between clusters in a t-SNE plot cannot reliably indicate the degree of similarity or dissimilarity. Also, choosing different values for perplexity and other parameters can lead to significantly different visualizations, further complicating the interpretation of global relationships.
